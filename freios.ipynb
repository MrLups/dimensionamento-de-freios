```python
import matplotlib.pyplot as plt
import os
import pandas as pd
if os.path.exists('./pasta'):
    print('Já existe')
else: 
    os.mkdir('./pasta')

Df = pd.DataFrame(columns=['distância', 'aceleração', 'Força Freio'])
c = ''
while c != 'N':  
    vf = 19.44
    vi = 22.22
    d = float(input('Distância: '))
    aceleração = (((vf*2)-(vi*2))/(2*d))
    print (f'Desaceleração: {aceleração*-1}')
    massa_carro = 850
    força_freio = ((massa_carro*(aceleração*-1))/9.81)
    print(f'Força freio: {força_freio}')
    c = str(input('Continuar? ')).capitalize()
    Df = Df.append(pd.Series( {"distância": d, "aceleração": (aceleração*-1), "Força Freio": força_freio}), ignore_index=True)
print(Df)
continuar = str(input('Deseja salvar o arquivo?[S] ou [N] ')).capitalize()
if continuar == 'S':
    pasta='./pasta'
    cont = 1
    for diretorio, subpastas, arquivos in os.walk(pasta):
        for arquivo in arquivos:
            cont = cont+1
            print(os.path.join(diretorio, arquivo))
            print(cont)
    writer = pd.ExcelWriter(f'./pasta/Dados de frenagem {cont}.xlsx')
    Df.to_excel(writer)
    writer.save()
    print("DataFrame is exported successfully to 'Arquivo_excell' Excel File.")
else:
    print('O programa foi encerrado sem salvar os dados!')
print('graficos')

df_novo= pd.read_excel(f'pasta\dados de frenagem {cont}.xlsx')
print(df_novo)



df_novo['distância', 'aceleração']
df_novo.plot(kind='line',figsize=(10,5))
plt.title('distância X aceleração')
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import math as m
import os
if os.path.exists('./pasta'):
    print('Já existe')
else: 
    os.mkdir('./pasta')


Df = pd.DataFrame(columns=['Força Pinça [N]','Força Fricção [N]' ,'Torque de frenagem [N/m]','Area Max Furada [m^2]', 'Num. furos'])
c = ''
while c != 'N':  
    
    RaioEf = 99.875 #esta em milimetros, precisa passar pra metros

    Pressure = float(input('Pressão Cilindro Mestre [Pa]: '))
    AreaCaliper = float(706.86) #não condiz com a realidade (5mm de raio) esta em milimetros, precisa passar pra metros
    AreaPastilha = float(3521.856) #esta em milimetros, precisa passar pra metros
    ForcaCaliper = 2*((Pressure*AreaCaliper)/AreaPastilha)
    print(f'Força Pinça: {ForcaCaliper}')

    mi = float(input('Coeficiente de atrito: '))
    ForcaFriccao = ForcaCaliper*mi
    print(f'Força Fricção: {ForcaFriccao}')

    TorqueFrenagem = ForcaFriccao*RaioEf
    print(f'Torque de frenagem: {TorqueFrenagem}')

    AreaPastilhaF = float(input('Area da Pastilha Furada [m**2]: '))
    AreaMaximaFurada = AreaPastilha-AreaPastilhaF
    print(f'Area Máxima que pode ser furada: {AreaMaximaFurada}')

    RaioFuro = 3 #não condiz com a realidade e esta em milimetros, precisa passar pra metros
    NumFuros = AreaMaximaFurada//(m.pi*(RaioFuro**2))
    print(f'Numero de furos que podem ser feitos: {NumFuros}')    
        
    c = str(input('Continuar? ')).capitalize()
    Df = Df.append(pd.Series( {'Força Pinça [N]': ForcaCaliper,'Força Fricção [N]': ForcaFriccao,'Torque de frenagem [N/m]':TorqueFrenagem,'Area Max Furada [m^2]': AreaMaximaFurada, 'Num. furos': NumFuros}), ignore_index=True)
print(Df)
continuar = str(input('Deseja salvar o arquivo?[S] ou [N] ')).capitalize()
if continuar == 'S':
    pasta='./pasta'
    cont = 1
    for diretorio, subpastas, arquivos in os.walk(pasta):
        for arquivo in arquivos:
            cont = cont+1
            print(os.path.join(diretorio, arquivo))
            print(cont)
    writer = pd.ExcelWriter(f'./pasta/Dados de frenagem teste {cont}.xlsx')
    Df.to_excel(writer)
    writer.save()
    print("DataFrame is exported successfully to 'Dados de frenagem teste' Excel File.")

    df_novo= pd.read_excel(f'pasta\dados de frenagem teste {cont}.xlsx')

print(df_novo)


df_novo.plot(kind='line',figsize=(10,5))
plt.title('distância X aceleração')
```
